```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CPI by Year.csv')
df.head()
```

```python
value_columns = [col for col in df.columns if col != 'Year']

plt.figure(figsize=(10, 6))

for col in value_columns:
    plt.plot(df['Year'], df[col], label=col)

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Year vs All Food Categories')
plt.legend()
plt.grid(True)

# Condense ticks: show every 5th year
plt.xticks(df['Year'][::5])

plt.tight_layout()
plt.show()
```

```python
from langchain.agents import AgentExecutor, Tool
from langchain.tools import PythonREPLTool
from langchain_openai import ChatOpenAI

# 5 Create the Agent
from langchain.agents.format_scratchpad.llm import format_to_openai_function_messages
from langchain_core.messages import HumanMessage

# 1 Load your dataset
df1 = pd.read_csv('CPI by Year.csv')

# 2 Define the Python REPL tool
python_tool = PythonREPLTool()

# 3 Define the LLM (with tool calling)
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
)

# 4 Define tool for the agent
tools = [
    Tool(
        name="python_repl",
        func=python_tool.run,
        description=(
            "Executes Python code. Use this to analyze the dataset and generate "
            "matplotlib charts. DataFrame is available as `df`."
        ),
    )
]

def run_visual_agent(user_prompt):
    """Send prompt  LLM  Python code  Matplotlib output."""
    messages = [
        HumanMessage(
            content=(
                "You are a data visualization agent.\n"
                "You have a pandas DataFrame called `df1`.\n"
                "When the user asks for a chart, write Python code using matplotlib.\n"
                "Show the plot using plt.show().\n\n"
                f"User prompt: {user_prompt}"
            )
        )
    ]

    # Convert to tool-use format
    function_messages = format_to_openai_function_messages(messages)

    result = llm.invoke(function_messages, tools=tools)

    return result

# 6 Example execution
if __name__ == "__main__":
    prompt = "Plot Year vs Food, F&V, Fats&Oils as a multi-line chart."
    output = run_visual_agent(prompt)
    print(output)
```
